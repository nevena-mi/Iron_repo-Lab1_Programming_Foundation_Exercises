*Motion sensors* in each room to detect intruders.  
*Temperature sensors* to prevent frozen pipes or fire hazards.  
*Door sensors* to track entry points.  
*Smoke detectors* for fire safety.  

security alerts:  
away modus - motion detector.  
away modus - door opener.  
Multiple sensors triggered simultaneously - (possible break-in).  

Safety Alerts:  
Temperature below 35°F - (frozen pipe risk).  
Temperature above 95°F - (equipment failure).  
Smoke detected - (fire risk).  

Comfort Notifications:  

Temperature outside comfort range (65-75°F) when home is in "home" mode.  
Unusual patterns (like a door left open for >5 minutes).  

###homeguard_system.ipynb###

```python
import random # to test/simulate sensors    
import datetime # to measure the duration of door opened
```

```python
class SecuritySystem:
    def __init__(self):
        self.mode = 'home'  # default mode is 'home'
        self.motion = False # nomovement
        self.door_open = False # door closed
        self.door_open_time = None # there is no opening time, door is closed
        self.smoke = False # no smoke detected
        self.temperature = 70 # default temperature in Fahrenheit

    def simulate_sensors(self):
        # Simulate motion detection
        self.motion = random.choice([True, False])
        
        # Simulate door status
        #self.door_open = random.choice([True, False])
        
        # Simulate smoke detection
        self.smoke = random.choice([True, False])
        
        # Simulate temperature reading
        self.temperature = random.randint(20, 110)  # Random temperature between 20 and 110
        
        new_door_state = random.choice([True, False])
        if not self.door_open and new_door_state:  # door was closed and now opened
            self.door_open_time = datetime.datetime.now()
        elif self.door_open and not new_door_state:  # door was opened and now closed
            self.door_open_time = None  # reset the door open time
        self.door_open = new_door_state

    def check_security_alerts(self):
        if self.mode != "away":
            return              #if not away mode, no security alerts are needed
        if self.motion and self.door_open:
            print("SECURITY ALERT: Possible break-in!!!")
        elif self.motion:
            print("SECURITY ALERT: Motion detected!")
        elif self.door_open:
            print("SECURITY ALERT: Door opened!")

    def check_safety_alerts(self):
        if self.smoke:
            print("🔥 FIRE ALERT!")
        if self.temperature < 35:
            print("❄ Frozen pipe risk!")
        if self.temperature > 95:
            print("🌡 High temperature warning!")

    def check_comfort_notifications(self):
        if self.mode != "home":
            return
        if self.temperature < 65 or self.temperature > 75:
            print("😊 Comfort notification: Temperature outside preferred range.")
        if self.door_open_time:
            time_open = datetime.datetime.now() - self.door_open_time
            if time_open > datetime.timedelta(minutes=5):
                print("🚪 Door has been open for more than 5 minutes.")

    def display_status(self):
        print("\n" + "=" * 40)
        print("      HOME SECURITY SYSTEM")
        print("=" * 40)

        print(f"Mode: {self.mode.upper()}")

        print(f"Motion detected: {self.motion}")
        print(f"Door open: {self.door_open}")
        print(f"Smoke detected: {self.smoke}")
        print(f"Temperature: {self.temperature}°F")

        if self.door_open_time:
            print(f"Door opened at: {self.door_open_time.strftime('%H:%M:%S')}")
        else:
            print("Door opened at: ---")

        print("=" * 40)  

         

```

```python
system = SecuritySystem()

mode = input("Choose mode (home/away): ").strip().lower()

if mode in ["home", "away"]:
    system.mode = mode
else:
    print("Invalid mode. Using HOME.")

system.simulate_sensors()

system.display_status()

system.check_security_alerts()

system.check_safety_alerts()

system.check_comfort_notifications()
```
